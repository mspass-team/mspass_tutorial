## Filtering Data with For Loops and If Statements


````python
magnitudes = [2.1, 5.3, 1.8, 6.7, 3.2, 4.9, 7.1, 2.5]

for mag in magnitudes:
    if mag > 4:
        print(mag, " Significant earthquake!")
````

### For Loops with Nested If Statements

| Piece | What it does |
|---|---|
| for item in list: | Visits each item in the list one at a time. The list is `magnitudes` |
| Loop variable | Temporary name for the current item (e.g. `mag`) |
| Nested if | A condition checked on every item during the loop |
| 1st indent level | Code that belongs to the `for` loop |
| 2nd indent level | Code that belongs to the `if` inside the loop |

```python
magnitudes = [2.1, 5.3, 1.8, 6.7, 3.2, 4.9, 7.1, 2.5]

for mag in magnitudes:        # loop through every item
    if mag > 4:               # filter condition
        print(mag, " Significant earthquake!")  # only prints if True
```

- **Note 1:** The loop variable (`mag`) is temporary — it is overwritten on each pass
- **Note 2:** This is what pandas `df[df["col"] > 4]` does internally; see below
- **Note 3:** Two levels of indentation are required one for the loop, one for the if
- **Note 4:** A colon is required after both `for` and `if`


````python
station = {
    "name": "ANMO",
    "latitude": 34.95,
    "longitude": -106.46,
    "elevation": 1820
}

for key, value in station.items():
    if type(value) == float:
        print(key, "-->", value, "(numeric coordinate)")
````

### Looping Over a Dictionary

| Piece | What it does |
|---|---|
| for key in dict: | Loops through keys only |
| for value in dict.values(): | Loops through values only |
| for key, value in dict.items(): | Loops through both at once |
| type(value) | Returns the data type of a value |
| Unpacking | Python splits each pair into two variables automatically |

```python
station = {
    "name": "ANMO",
    "latitude": 34.95,
    "longitude": -106.46,
    "elevation": 1820
}

# Loop through keys and values at the same time
for key, value in station.items():
    if type(value) == float:          # filter by data type
        print(key, "-->", value, "(numeric coordinate)")

# Three ways to loop over a dictionary
for key in station:                   # keys only
for value in station.values():        # values only
for key, value in station.items():    # both at once, most commonly used

# Common types to check for
type(value) == str     # text
type(value) == int     # whole number
type(value) == float   # decimal number
type(value) == bool    # True or False
```

- **Note 1:** `.items()` gives both key and value and is used most often
- **Note 2:** `type()` is used for understanding mixed data or filtering by kind
- **Note 3:** Use `==` for comparison, not `=`

<a id="list-comprehension"></a>
### List Comprehension

**Syntax:**
```python
[expression for item in iterable if condition]
```

| Piece | What it does |
|---|---|
| `expression` | Value to collect into the new list |
| `for item in iterable` | Iterates over each element in the source |
| `if condition` | Optional filter — excludes elements that evaluate to `False` |

```python
magnitudes = [2.1, 5.3, 1.8, 6.7, 3.2]

# Filter returns elements meeting the condition
significant = [mag for mag in magnitudes if mag > 4]
# Returns: [5.3, 6.7]

# Applies expression to every element
rounded = [round(mag) for mag in magnitudes]
# Returns: [2, 5, 2, 7, 3]

# Equivalent for loop
significant = []
for mag in magnitudes:
    if mag > 4:
        significant.append(mag)
```


- **Note 1:** `expression` must appear first inside the brackets
- **Note 2:** The source `iterable` is not modified a new list is always returned
- **Note 3:** For complex logic, a standard `for` loop is preferred for readability

### Vectorized Operations

Unlike lists:
- Arrays store data in **contiguous memory**
- Operations are implemented in **compiled code** for speed

For example:

This is called **vectorization**.

#### Caesar Cipher

*(Source: http://www.ling.gu.se/~lager/python_exercises.html)*

In cryptography, a Caesar cipher is a simple encryption technique in which each letter in the plaintext is replaced by a letter a fixed number of positions down the alphabet. For example, with a shift of 3, A would be replaced by D, B would become E, and so on. The method is named after Julius Caesar, who used it to communicate with his generals. ROT-13 ("rotate by 13 places") is a widely used example of a Caesar cipher where the shift is 13. In Python, the key for ROT-13 can be represented with the following dictionary:

```python
key = {'a':'n', 'b':'o', 'c':'p', 'd':'q', 'e':'r', 'f':'s', 'g':'t', 'h':'u',
       'i':'v', 'j':'w', 'k':'x', 'l':'y', 'm':'z', 'n':'a', 'o':'b', 'p':'c',
       'q':'d', 'r':'e', 's':'f', 't':'g', 'u':'h', 'v':'i', 'w':'j', 'x':'k',
       'y':'l', 'z':'m', 'A':'N', 'B':'O', 'C':'P', 'D':'Q', 'E':'R', 'F':'S',
       'G':'T', 'H':'U', 'I':'V', 'J':'W', 'K':'X', 'L':'Y', 'M':'Z', 'N':'A',
       'O':'B', 'P':'C', 'Q':'D', 'R':'E', 'S':'F', 'T':'G', 'U':'H', 'V':'I',
       'W':'J', 'X':'K', 'Y':'L', 'Z':'M'}
```

The exercise is to implement a decoder for ROT-13. Once implemented, use it to decode the following message:

```

```
