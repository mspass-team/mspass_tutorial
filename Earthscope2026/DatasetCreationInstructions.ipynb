## Overview
Earthscope is in the process of moving their entire seismic waveform archive to Amazon's S3 storage.   Soon all data requests for waveform data will come from that archive.  The data set you will assemble for this course takes advantage of that new capability. It also teaches a very important, fundamental lesson on how you should aim to do your work in the future.   That lesson is *reproducible science*.   You will hear me preach on this in the opening session lecture, but there has been a huge issue in any science like seismology that depends on computing of irreproducible results.  Many studies use large data sets that are impossible for someone else to recreate exactly let alone validate the code they ran for something like a tomographic inversion.   There is an important movement in all science to address this issue.   I assert one of the most important lessons you can get from this course is a recipe for reproducible assembly of the data you use for your research.   That lesson is the sequence of notebooks you will run in this course that begin with the raw archive and end with a processed result; in this case stacked P wave signals from a set of teleseismic events.  

## Step 1:  start database server
Throughout this course we will be using a database system called MongoDB.   You will learn a lot more about this in Session 1, but for now the important thing is you will need to launch a private copy of a MongoDB server for your use.   None of the notebooks in this class will work without that server running.   Do the following:

1.  Use the jupyter launcher directly or hit the + tab create a launcher to create a "Terminal" winow.  Select the "Terminal" tab in the GeoLab window where you are reading this and you should see a prompt form the unix shell.  
2.  Select that the tab you created.  You should see a unix shell prompt.  Run the following shell incantation in that window  (you can cut and paste the two lines below):
```
#  makes sure you are running this in your home directory
cd
# these are needed only the first time use GeoLab with MsPASS
mkdir db
mkdir logs
# this launches mongodb.    Note it references "db" and "logs"
mongod --dbpath ./db --logpath ./logs/mongo_log
```
Note:
-  The command will block.  (i.e. the Terminal window will no longer accept input after the mongod command is issued) Do not try to enter anything else in that terminal window.  
-  DO NOT exit the terminal window from which you run the mongod command or you will kill the database server.  
-  Each time you reconnect to GeoLab you will need to relaunch the MonogDB server with that  incantation. Be careful you always run the command from the same directory as that way the database files will be written to ~/db and the mongodb log will appear in ~/logs.  If you run that command from a different location anything you did previously won't be lost but will just not be visible.  It is similar to being in the wrong directory looking for files that are somewhere else. 
-  In all cases it is wise to launch a second Terminal window and verify that worked by typing `ps -A`  You should see a line where the CMD field is "mongod".   If not, contact me by email or slack if you are unable to solve the problem.  The first place to look if you are having a problem is the content of the file ~/logs/mongo_log that will be generated when this command runs.

## Step 2: load arrival time data
You first need to run the notebook with the file name **load_antelope.ipynb**. I recommend you run that now and while it is running read the following brief paragraph.

The data you are loading are relational database tables created by the Earthscope ANF.   The tables contain arrival time picks for March of 2010.   They are a fraction of the available data that could be handled by this mechanism.   Tar files of monthly catalog data in this format are available [here](https://doi.org/10.17611/dp/eb.1).  I say that because with this template it would be relatively easy to extend this processing to the entire usarray data set with ANF picks.   There is, however, one caveat.  You will find that two files you need to make this work are not on that site but are something I got from Frank Vernon who headed the ANF operation for the TA.   Those are two files "css30/usarray_2010_03.snetsta" and "css30/usarray_2010_03.schanloc".   Those tables are actually valid for the entire TA recording period, not just March 2010 as the name suggests.   Unless you want to build on this work after the course that is an irrelevant detail but important to document here as it would be required to truly make this a reproducible workflow.  

Loading the arrival time data is very quick.   The thing this notebook also does is use obspy's web service gizmo to assemble source and receiver data.   It then loads that data into MongoDB.   That process takes a bit longer so be patient.   We'll look at that data at length in our first class session.
